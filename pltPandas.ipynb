```python
import pandas as pd
df = pd.read_csv('C:\\WORK\\mySQLstudy\\titanic.csv')
```

```python
df.head()
```

```python
%matplotlib inline
df['Fare'].plot.hist(bins=30)
```

```python
df.Fare.plot.kde()
```

```python
df.plot.scatter(x='Fare', y='Survived')
```

```python
import matplotlib.pyplot as plt
df.groupby('Survived').Fare.plot.kde()
plt.xlim(-10,200)
plt.legend()
```

```python
ax=df['Fare'].plot.hist()
ax.set_title('visualization')
ax.figure.savefig('titanic.pdf')
```

```python
fig, ax = plt.subplots(figsize=(10,5))
df.Survived.plot.kde(label='All', ax=ax)
for label, class_df in df.groupby('Pclass'):
    print(label)
    class_df.Survived.plot.kde(ax=ax, label=label)
plt.legend()
```
